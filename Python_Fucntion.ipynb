#Q1. What is the difference between a function and a method in Python?

* A `function` is a piece of code you write once and use again and again wherever needed. It is not part of any object.
          def sum(a, b):
            return a + b

Now you can use sum(2, 3) anywhere in your code to get the sum of two numbers.

* A `method` is also a function, but it is connected to an object like a string, list, or class. You use it with dot (.) after the object.

      name = "Yash"
        print(name.upper())  

So, a method is just a function that belongs to an object.

#Q3. What are the different ways to define and call a function in Python?

Ways to define a function:

* Using `def` (normal function)

        def greet(name):
          return f"Hello, {name}!"
  
* Using `lambda` (anonymous function)
   
      greet = lambda name: f"Hello, {name}!"
  

* `Nested function` (function inside another function)
  
      def outer():
       def inner():
           return "Inner function"
       return inner()
   

* `Function with default argument`
   
      def greet(name="Yash"):
        return f"Hello, {name}!"
   

* Function with `*args` and `**kwargs` (When you don’t know how many values will be passed)

        def show_args(*args, **kwargs):
          print("Positional:", args)
          print("Keyword:", kwargs)


Ways to **call** a function:

* `Normal call`
   
      greet("Rahul")
   

* `Using positional and keyword arguments`

      show_args(1, 2, name="Ajay", age=30)
   

* `Calling a lambda function`

      (lambda x: x * 2)(5)  
   

* `Calling a function returned from another function`
   
      def outer():
        def inner():
           return "Inner Function"
       return inner

   inner_func = outer()
   print(inner_func())

#Q5. What are iterators in Python and how do they differ from iterables?

* Iterable: Any object you can loop over (like list, tuple, string).
* Iterator: An object that remembers its position and gives the next value when asked.

Difference:

* Iterable needs to be converted to an iterator using `iter()`.
* Iterator uses `next()` to get values one by one.

      nums = [1, 2, 3]        # iterable
      it = iter(nums)         # iterator
      print(next(it))         # 1
      print(next(it))         # 2



#Q6. Explain the concept of generators in Python and how they are defined?

* Generators are special functions that return values **one at a time** using the `yield` keyword.
* They remember their position, so they can continue from where they left off.
* They are lazy meaning that values are generated only when needed.
* They use less memory because they don't store all values at once.

How to define a generator:

      def count(n):
        i = 1
        while i <= n:
          yield i
          i += 1

    for num in count(3):  
    print(num, end = " ")    #Output 1 2 3




#Q7. What are the advantages of using generators over regular functions?

* Uses less memory --> It doesn't create the full list in memory. It gives one value at a time.
* Lazy evaluation --> It gives the next value only when you ask for it.
* Faster for large data --> Since it doesn't make a full list, it works faster with big data.
* Good for infinite sequences --> You can keep getting values one by one without stopping (like in Fibonacci), and it won't crash the system.


#Q8. What is a lambda function in Python and when is it typically used?

* A lambda function is a small, one-line anonymous function (without a name).

* It's used for short tasks where a full def function is not needed — like inside map(), filter(), or sort().

      num = [1, 2, 3, 4]
      sqr = list(map(lambda x: x**2, num))
      print(sqr)  # Output: [1, 4, 9, 16]









#Q9. Explain the purpose and usage of the `map()` function in Python.

* The map() function is used to apply a function to each item in a list (or any iterable) and return a new result.

* The purpose is to process all items in a list without using a loop.

      num = [1, 2, 3, 4]
      sqr = list(map(lambda x: x**2, num))
      print(sqr)  # Output: [1, 4, 9, 16]



#Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

* `map()` Applies a function to each item and returns a new list.

      map(lambda x: x*2, [1, 2, 3]) → [2, 4, 6]`

* `filter()` Keeps only items that match a condition.

      filter(lambda x: x > 2, [1, 2, 3]) → [3]

* `reduce()` Repeatedly applies a function to items and reduces them to a single value. Needs to import from `functools`

      `reduce(lambda x, y: x + y, [1, 2, 3]) → 6`

```python
'''1. Write a Python function that takes a list of numbers as input
and returns the sum of all even numbers in the list'''

def add(l):
  a = 0
  for i in l:
    if(i % 2 == 0):
      a = a + i
  return a

l1 = list(map(int, input("Enter number with spaces: ").split()))
z = add(l1)
print(z)
```

```python
'''4. Write a Python function that checks if a given number is prime or not
from 1 to 200.'''

def check_prime(l4):
    prime = []
    for i in range(2, l4+1):
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            prime.append(i)
    return prime

p = check_prime(200)
print(p)
```

```python
''' 5. Create an iterator class in Python that generates the Fibonacci sequence
up to a specified number of terms '''

class Fibonacci_series:

  def __init__(self,num):
    self.num = num
    self.a = 0
    self.b = 1
    self.count = 0

  def __iter__(self):
        return self

  def __next__(self):

    for i in range (self.num):
      if self.count >= self.num:
        raise StopIteration
      result =  self.a
      self.a , self.b = self.b, self.a + self.b
      self.count += 1
      return result



c = Fibonacci_series(10)
print(next(c),next(c),next(c),next(c),next(c),next(c),next(c),next(c),next(c))



```

```python
'''6. Write a generator function in Python that yields the powers of 2 up to a
given exponent '''


def expo(l):
  for i in range (0,l):
    yield (2**i)

num = expo(5)
print(next(num))
```

```python
'''8. Use a lambda function in Python to sort a list of tuples based on the
second element of each tuple.'''

def sort():
  tup = ((1,3),(1,2),(1,1),(1,5))
  a = sorted(tup, key = lambda x: x[1])
  return a

sort()
```

```python
'''9. Write a Python program that uses `map()` to convert a list of temperatures
from Celsius to Fahrenheit.'''

def convert():
  l1 = [32,42,38]
  temp = list(map(lambda x: round(9/5 * x + 32,2),l1))
  return temp

convert()
```

```python
'''10. Create a Python program that uses `filter()` to remove all
the vowels from a given string.'''

def fil():
  data = input("Enter String: ")
  s = list(filter(lambda x: False if x.lower() in ["a","e","i","o","u"]
                  else True, data))
  print(''.join(s))

fil()
```

```python
'''11) Imagine an accounting routine used in a book shop. It works on a list
with sublists, which look like this:'''

''' Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map'''


order_no = tuple(map(int, input("Enter order numbers with space: ").split()))
price_per_item = tuple(map(int, input("Enter price per item with space: ").split()))
quantity_no = tuple(map(int, input("Enter quantity numbers with space: ").split()))

total_price = list(map(lambda x: x[0] * x[1], zip(price_per_item, quantity_no)))
final_price = list(map(lambda x: x + 10 if x < 100 else x, total_price))

result = list(map(lambda i: (order_no[i], final_price[i]), range(0,len(order_no))))

print(result)

```
